#To analyze the relationships between variables in a dataset, generating a correlation matrix is a common and effective approach.between variables like **Skinfirmness, PhysiologicalMaturity, Bunch Length (BL), Bunch Weight (BW), and Seed Weight**.

```python
import pandas as pd
data = pd.DataFrame({
    'Skinfirmness': [60.9, 40.7, 51.2, 38.2, 52.6],
    'PhysiologicalMaturity': [127.0, 167.66, 141.66, 121.33, 127.33],
    'BunchLength': [18.23, 17.66, 23.66, 17.60, 21.26],
    'BunchWeight': [104.80, 182.77, 176.86, 137.36, 237.71],
    'SeedWeight': [4.20, 11.83, 5.39, 4.03, 3.46]
})
correlation_matrix = data.corr()
print(correlation_matrix)
```

#visualize the corelation using heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix of Variables")
plt.show()

```

#Visualized boxplots for each variable to highlight outliers.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    'ABW': [75.0,263.03,134.43,34.66,154.96],
    'BW': [104.80, 182.77, 176.86, 137.36, 237.71],
    'SeedWeight': [4.20, 11.83, 5.39, 4.03, 3.46]
})
plt.figure(figsize=(15, 5))
# Boxplot for ABW
plt.subplot(1, 3, 1)
sns.boxplot(x=df['ABW'])
plt.title('Boxplot of ABW')
# Boxplot for BW
plt.subplot(1, 3, 2)
sns.boxplot(x=df['BW'])
plt.title('Boxplot of BW')
# Boxplot for SeedWeight
plt.subplot(1, 3, 3)
sns.boxplot(x=df['SeedWeight'])
plt.title('Boxplot of SeedWeight')
plt.tight_layout()
plt.show()
```

#Histograms to analyze the distribution of numerical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.DataFrame({
    'ABW': [75.0,263.03,134.43,34.66,154.96],
    'BW': [104.80, 182.77, 176.86, 137.36, 237.71],
    'SeedWeight': [4.20, 11.83, 5.39, 4.03, 3.46]
})
plt.figure(figsize=(15, 5))
# ABW Histogram
plt.subplot(1, 3, 1)
sns.histplot(data['ABW'], kde=True, color='skyblue', bins=10)
plt.title('Histogram of ABW')
plt.xlabel('ABW')
# BW Histogram
plt.subplot(1, 3, 2)
sns.histplot(data['BW'], kde=True, color='orange', bins=10)
plt.title('Histogram of BW')
plt.xlabel('BW')
# SeedWeight Histogram
plt.subplot(1, 3, 3)
sns.histplot(data['SeedWeight'], kde=True, color='green', bins=10)
plt.title('Histogram of SeedWeight')
plt.xlabel('SeedWeight')
plt.tight_layout()
plt.show()
```

#Scatter plots to explore relationships between key variables.

```python

plt.figure(figsize=(15, 5))
# Scatter plot for ABW vs. BW
plt.subplot(1, 3, 1)
sns.scatterplot(x=data['ABW'], y=data['BW'], color='blue')
plt.title('Scatter Plot of ABW vs. BW')
plt.xlabel('ABW')
plt.ylabel('BW')
# Scatter plot for ABW vs. SeedWeight
plt.subplot(1, 3, 2)
sns.scatterplot(x=data['ABW'], y=data['SeedWeight'], color='red')
plt.title('Scatter Plot of ABW vs. SeedWeight')
plt.xlabel('ABW')
plt.ylabel('SeedWeight')
# Scatter plot for BW vs. SeedWeight
plt.subplot(1, 3, 3)
sns.scatterplot(x=data['BW'], y=data['SeedWeight'], color='green')
plt.title('Scatter Plot of BW vs. SeedWeight')
plt.xlabel('BW')
plt.ylabel('SeedWeight')
plt.tight_layout()
plt.show()
```
