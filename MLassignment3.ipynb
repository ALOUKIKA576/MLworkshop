##1. Basic Line Plot
o Plot a simple line graph of ùë¶ = ùë•^2

for ùë• values ranging from -10 to 10.

o Label the x-axis as "X Values" and the y-axis as "Squared Values."

o Add a title to the graph.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x, y)
plt.xlabel('X Values')
plt.ylabel('Squared Values')
plt.title('Line Plot of y = x^2')
plt.grid(True)
plt.show()
```

#3.Scatter Plot
o Generate two lists of 20 random integers each (for x and y coordinates).

o Create a scatter plot with these values, and customize it by adding color,
size, and labels to the points.


```python
import matplotlib.pyplot as plt
import numpy as np

# Generate two lists of 20 random integers each for x and y coordinates
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)
# Create a scatter plot
plt.scatter(x, y, color='green', s=100, edgecolor='black', alpha=0.7)
# labeling the points
for i in range(len(x)):
    plt.text(x[i] + 1, y[i] + 1, f'({x[i]}, {y[i]})', fontsize=9)
#labling the axis
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
#title for the plot
plt.title('Random Scatter Plot with Customization')
plt.grid(True)
plt.show()

```

##4. Histogram
o Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.

o Customize the number of bins and add a grid to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Random Values from Normal Distribution')
plt.grid(True)
plt.show()
```

##5. Pie Chart
o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].

o Display the percentage values on the chart and use a color palette for
differentiation.

```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [20, 30, 25, 15, 10]
#colors for piechart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
# Add a title
plt.title('Percentage Distribution of Categories')
plt.show()


```

##6. Box Plot (Seaborn)

##7. Heatmap (Seaborn)
o Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.

o Add a color bar and annotate the heatmap with the values.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
# to  generate 5x5 matrix
matrix = np.random.randint(1, 11, size=(5, 5))
# to create a heatmap
sns.heatmap(matrix, annot=True, fmt='d', cmap='YlGnBu', cbar=True)
# Add title
plt.title('5x5 Heatmap with Random Integers')
plt.show()
```

##8. Comparison of Matplotlib and Seaborn
o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.

o Highlight the differences in the code, syntax, and aesthetics of the two
libraries

```python
import matplotlib.pyplot as plt
import numpy as np
students = ['Alice', 'Bob', 'Carol', 'David', 'Eve']
subjects = ['Math', 'Science', 'English']
scores = np.random.randint(50, 100, (5, 3))
fig, ax = plt.subplots(figsize=(8, 6))
width = 0.2
x = np.arange(len(students))

for i, subject in enumerate(subjects):
    ax.bar(x + i * width, scores[:, i], width, label=subject)
ax.set_xlabel('Students')
ax.set_ylabel('Scores')
ax.set_title('Student Scores by Subject (Matplotlib)')
ax.set_xticks(x + width)
ax.set_xticklabels(students)
ax.legend()

# Display the plot
plt.show()
```

##10. Save and Style Plots
‚Ä¢ Plot any graph of your choice and save it in both .png and .pdf formats.

‚Ä¢ Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain
how the style impacts the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
x = np.linspace(-10, 10, 100)
# Generate 100 points from -10 to 10
y = x**2
# Apply the style
plt.style.use('ggplot')
# Plotting the graph
plt.plot(x, y, label='$y = x^2$', color='blue', linewidth=2)
# Add labels and title
plt.xlabel('X Values')
plt.ylabel('Squared Values')
plt.title('Line Plot of $y = x^2$')
# Adding a legend
plt.legend()
plt.savefig('plot.png')
plt.savefig('plot.pdf')
plt.show()

```
